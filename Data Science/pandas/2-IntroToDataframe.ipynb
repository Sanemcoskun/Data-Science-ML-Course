```python
# multi index 
```

```python
first_index = ["Simpson","Simpson","Simpson","South Park","South Park","South Park"]
```

```python
inner_index = ["Homer","Bart","Marge","Cartman","Kenny","Kyle"]
```

```python
zip_index = list(zip(first_index,inner_index))
```

```python
zip_index
```

```python
zip_index = pd.MultiIndex.from_tuples(zip_index)
```

```python
zip_index
```

```python
sample_values = np.ones((6,2))
```

```python
big_df = pd.DataFrame(sample_values, index=zip_index, columns=["age","salary"])
```

```python
big_df
```

```python
big_df["age"]
```

```python
big_df.loc["Simpson"].loc["Homer"]
```
