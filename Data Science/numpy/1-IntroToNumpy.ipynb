```python
# dizi boyutları aynı olmalı yoksa hata verir
```

Broadcasting, farklı boyutlardaki (shape’leri farklı) dizilerin (array) aritmetik işlemler sırasında otomatik olarak aynı boyuta genişletilmesi işlemidir.Normalde iki dizinin toplama, çıkarma gibi işlemlerde kullanılabilmesi için boyutlarının (shape) aynı olması gerekir.
Ancak broadcasting sayesinde — belli kurallara uyuyorsa — küçük dizi otomatik olarak büyük dizinin boyutuna “yayılır” (broadcast edilir).
